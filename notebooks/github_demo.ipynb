```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline 
```

```python
data = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range('1/1/2000', periods=1000), columns=list('ABCD'))
random_walk = data.cumsum()
bars = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
bars = bars.cumsum()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))
random_walk.plot(kind='line', ax=axes[0], title='line')
bars.plot(kind='bar', ax=axes[1], title='bars')
```
