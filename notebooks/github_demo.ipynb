```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline 
```

```python
scatter = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range('1/1/2000', periods=1000), columns=list('ABCD'))
random_walk = scatter.cumsum()
scatter = scatter*(scatter.ix[:,0]+5)*(scatter.ix[0,:]+5)
bars = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
bars = bars.cumsum()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20,5))
random_walk.plot(kind='line', ax=axes[0], title='line')
bars.plot(kind='bar', ax=axes[1], title='bars')
scatter.plot('A','B',kind='scatter', ax=axes[2], title='scatter')
axes[2].set_xlabel('')
axes[2].set_ylabel('')
```
